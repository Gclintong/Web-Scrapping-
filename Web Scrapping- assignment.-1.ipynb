```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url ="https://quotes.toscrape.com/page/1/"
```

```python
response = requests.get(url)
```

```python
response = response.content
```

```python
soup = BeautifulSoup(response,'html.parser')
```

```python
quote_list = soup.find_all('div', {'class': 'quote'})
```

```python
for quote in quote_list:
    text = quote.find('span', {'class': 'text'}).text
    print(text)
```

```python
for quote in quote_list:
    tags = [tag.text for tag in quote.find_all('a', {'class': 'tag'})]
    print('Tags:', tags)
```

```python
for quote in quote_list:
    author = quote.find('small', {'class': 'author'}).text
    print('Author:', author)
```

```python
quotes = []
for quote in quote_list:
    text = quote.find('span', {'class': 'text'}).text
    tags = [tag.text for tag in quote.find_all('a', {'class': 'tag'})]
    author = quote.find('small', {'class': 'author'}).text
    quotes.append([text, tags, author])
    print(text,   tags  , author)
```

```python
quotes = []
for i in range(1,4):
    url ="https://quotes.toscrape.com/page/{}/".format(i)
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response,'html.parser')
    quote_list = soup.find_all('div', {'class': 'quote'})
    for quote in quote_list:
        text = quote.find('span', {'class': 'text'}).text
        tags = [tag.text for tag in quote.find_all('a', {'class': 'tag'})]
        author = quote.find('small', {'class': 'author'}).text
        quotes.append([text, tags, author])
```

```python
df=pd.DataFrame(quotes,columns = ["text", "tags", "author"] )
```
